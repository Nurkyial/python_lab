```python
# скачиваем статью и сохраняем его в txt файл
from bs4 import BeautifulSoup
import requests

r = requests.get("https://ru.wikipedia.org/wiki/Python")

def request_html(r):
  html = r.text
  #print(r.status_code)
  soup = BeautifulSoup(html, 'html.parser')
  return soup.body.get_text().strip()

with open("article.txt", 'w') as f:
  f.write(request_html(r))
```

```python
def read_file_as_dict(file_name): # функция которая считает количество уникальных слов в статье и выводит как словарь
  words = set()
  lst = []
  with open(file_name, 'r') as file:
    for line in file.read().split('\n'):
        for word in line.split():
            words.add(word)
            lst.append(word)

  result = {}
  for word in words:
    result[word] = lst.count(word)

  return result

dict_of_words = read_file_as_dict('article.txt')
print(dict_of_words)
```

5.	Создайте функцию, которая принимает список слов и возвращает словарь, где ключи — это слова, а значения — это множества всех слов, которые появляются после данного слова в тексте. Например, если текст содержит фразу “я люблю яблоки и я люблю груши”, то для ключа “люблю” значение будет {"яблоки", "груши"}.

```python
def create_word_dict(words):
    word_dict = {}
    for i in range(len(words)):
        word = words[i]
        if i == len(words) - 1:
            word_dict[word] = set()
            continue
        word_dict[word] = {words[j] for j in range(i+1, len(words))}
    return word_dict

text = 'Я люблю яблоки и груши'
words = text.split()
word_dict = create_word_dict(words)
print(word_dict)
# {'люблю': {'яблоки', 'груши'}}
```

6.	Создайте функцию, которая принимает словарь (созданный на предыдущем шаге) и строку, и выводит все возможные продолжения строки на основе данных в словаре. Например, если строка “я люблю”, то функция может вывести “яблоки” или “груши”.


```python
def generate_continuations(word_dict, prefix):
    words = word_dict[prefix]
    if words:
        for word in words:
            print(word)
    else:
        print("Word not found in dictionary")

print(generate_continuations(word_dict, 'Я'))
```

•	Запишите результаты в таблицу.


```python
from timeit import timeit
import sys

def analyze_data_type(data_type, words):
    conversion_time = 0
    mem_usage = 0

    #def data_type():

    def convert_to_data_type():
        nonlocal conversion_time
        nonlocal mem_usage

        result = data_type()
        if result == list():
          #print('yes')
          for word in words:
            result.append(word)
        elif result == set():
          for word in words:
            result.add(word)
        elif result == dict():
          for word in words:
            result[word] = result.get(word, 0) + 1
          # or other methods depending on data type
        return result

    t = timeit(convert_to_data_type)
    mem = sys.getsizeof(convert_to_data_type)

    return f"{data_type.__name__}: {t:.3f}s, {mem:.0f}b"

your_list_of_words = ['apple', 'banana', 'cherry']

data_types = ['list', 'set', 'dict']
for dt in data_types:
    print(analyze_data_type(eval(dt), your_list_of_words))
```
